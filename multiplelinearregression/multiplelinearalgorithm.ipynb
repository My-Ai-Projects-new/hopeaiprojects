```python
import pandas as pd
```

```python
dataset=pd.read_csv("multiple_linear_regression_30_rows.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent= dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_California', 'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
dependent
```

```python
#train model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#learn 
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
#evaluation
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename="finalized_model_mul_linear.sav"
pickle.dump(regressor,open(filename,"wb"))
```
