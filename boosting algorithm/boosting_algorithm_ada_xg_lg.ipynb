```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

from xgboost import XGBRegressor
```

```python
df = pd.read_csv("insurance_pre.csv")
```

```python
df
```

```python
df = pd.get_dummies(df, drop_first=True).astype(int)
```

```python
df
```

```python
X = df.drop("charges", axis=1)
y = df["charges"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
ada = AdaBoostRegressor()
ada.fit(X_train, y_train)

ada_pred = ada.predict(X_test)
ada_r2 = r2_score(y_test, ada_pred)

print("AdaBoost R2 Score:", ada_r2)
```

```python
gb = GradientBoostingRegressor()
gb.fit(X_train, y_train)

gb_pred = gb.predict(X_test)
gb_r2 = r2_score(y_test, gb_pred)

print("Gradient Boosting R2 Score:", gb_r2)
```

```python

xgb = XGBRegressor()
xgb.fit(X_train, y_train)

xgb_pred = xgb.predict(X_test)
xgb_r2 = r2_score(y_test, xgb_pred)

print("XGBoost R2 Score:", xgb_r2)
```

```python
print("\nModel Comparison:")
print("AdaBoost:", ada_r2)
print("Gradient Boosting:", gb_r2)
print("XGBoost:", xgb_r2)
```
